### R Data Visualizations with ggplot2


```r
# get the library
library(ggplot2)
# grammar of graphics with layers concept
#1. data layer -- On what data to perform the plot
#2. aesthetics layer -- what are the plot elements
#3. geometries layer -- Type of plot 
#4. Facets
#5. Statistics
#6. coordinates ( Theme layer)

```

```r
head(mtcars,3)
```

```r
plt <- ggplot(data = mtcars, aes(x = mpg, y= hp))
plt <- plt+geom_point() # this is for the scatter plot
```

```r
plt # we can see a sort of co-relation here higher the hp, lesser the milage.
```

```r
# Adding the other 3 layers one by one and see the effect
```

```r
plt <- plt + facet_grid(cyl~.)
plt <- plt+stat_smooth()
plt
```

####  Histogram frequency plot

```r
plt <- ggplot(data = mtcars, aes(x = mpg))
## Add layers 
plt <- plt+ geom_histogram(binwidth =5, color = 'red', fill = 'blue', alpha = 0.3)
#Adding lables for x an and y
plt <- plt+ xlab('Mileage of cars') + ylab('Frequency')
# Adding title to the plot
plt <- plt + ggtitle(" Mileage Histogram")
plt
```
